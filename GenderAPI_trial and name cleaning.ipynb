```python
## example of using gender API:

import json
from urllib.request import urlopen

def get_gender(names, myKey):
    results = {}
    base_url = "https://gender-api.com/get?key=" + myKey
    for name in names:
        url = f"{base_url}&name={name}"
        response = urlopen(url)
        decoded = response.read().decode('utf-8')
        data = json.loads(decoded)
        results[name] = data["gender"]  # Storing the gender result
        print(f"Name: {name}, Gender: {data['gender']}")

    return results

# Example usage
myKey = "fccbe5541663308e7568cbc916144090a2a55df706849ea2506c7758bd37c1f6"
names = ["kevin", "markus", "lisa"]
gender_results = get_gender(names, myKey)
```

```python
import pandas as pd
import json
from urllib.request import urlopen, quote

def get_gender_api(name, myKey):
    if pd.notnull(name):
        first_name = name.split()[0]  # Extracting the first name
        encoded_name = quote(first_name.encode('utf-8'))  # Encoding the name to handle special characters
        url = f"https://gender-api.com/get?key={myKey}&name={encoded_name}"
        try:
            response = urlopen(url)
            decoded = response.read().decode('utf-8')
            data = json.loads(decoded)
            return data.get('gender', 'unknown')
        except Exception as e:
            print(f"Request to Gender API failed: {e}")
            return 'unknown'
    else:
        return 'unknown'

# Example usage
myKey = "insert your server key here"  # Replace this with your actual Gender API key
participants_links_expanded['gender_api'] = participants_links_expanded['name'].apply(lambda x: get_gender_api(x, myKey))

# Display the first few entries to verify
print(participants_links_expanded[['name', 'gender_api']])
```

```python
import pandas as pd
import re
import unicodedata

def clean_names(column):
    # Define regex patterns
    title_suffix_pattern = r'\b(Dr\.|Mr\.|Mrs\.|Ms\.|Jr\.|Sr\.|III|IV|PhD|MD)\b'
    numbers_pattern = r'[0-9]+'
    
    def clean_name(name):
        # Normalize international characters
        name = unicodedata.normalize('NFKD', name).encode('ascii', 'ignore').decode('utf-8')
        
        # Remove titles and suffixes
        name = re.sub(title_suffix_pattern, '', name)
        
        # Insert space before capital letters (but not at the beginning) and remove numbers
        name = re.sub(r'(?<!^)(?=[A-Z])', ' ', name)
        name = re.sub(numbers_pattern, '', name)
        
        # Remove extra spaces
        name = re.sub(r'\s+', ' ', name).strip()
        
        # Convert to title case
        name = name.title()
        
        return name
    
    return column.apply(clean_name)

# Assuming 'participants_links_expanded' is your DataFrame and 'name' is the column to be cleaned
participants_links_expanded['cleaned_names'] = clean_names(participants_links_expanded['name'])

# Correct way to display two columns from the DataFrame
print(participants_links_expanded[['name', 'cleaned_names']])
```
